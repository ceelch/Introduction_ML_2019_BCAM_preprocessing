# <font color='orange'> * Outliers detection * </font>

## As was mentioned in the slides, the ourliers values are outside of the following interval:

$$ [Q_1-1,5*IQR, Q_3+1.5*IQR]$$

## The following boxplot figure shows this method

## Lets plot the boxplot for each variable

```python
ax = sns.boxplot(data=data1[data1.columns[:-1]], orient="h", palette="Set2", linewidth=2.5)
plt.show()
```

```python
ax2 = sns.boxplot(y="sepal_width", data=data1, orient="h", color=sns.color_palette("Set2")[1], linewidth=2.5)
plt.show()
```

```python
IQR = data1_statistic["sepal_width"]["75%"] - data1_statistic["sepal_width"]["25%"]
whiskers = [data1_statistic["sepal_width"]["25%"]-(1.5*IQR), data1_statistic["sepal_width"]["75%"]+(1.5*IQR)]
```

```python
yvector_0=whiskers[0]*np.ones(len(data1))
yvector_1=whiskers[1]*np.ones(len(data1))
```

```python
fig =plt.figure(figsize=(20, 6))
sns.scatterplot(x=data1.index, y='sepal_width', data=data1, color=sns.color_palette("Set2")[1],s=50);
plt.xlabel('index',fontsize=22)
plt.ylabel('width',fontsize=22)
plt.title('Sepal Width',fontsize=22)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)

sns.lineplot(x=data1.index,y=yvector_0,dashes=False, lw=3, color=sns.color_palette("Blues")[4])
sns.lineplot(x=data1.index,y=yvector_1,dashes=False, lw=3, color=sns.color_palette("Blues")[4])

sns.despine()
plt.show()
```

```python
index_outliers=data1.index[(data1['sepal_width'] > whiskers[1]) | (data1['sepal_width'] < whiskers[0])].tolist()
```

```python
index_outliers
```

```python
data1_outliers=data1.drop(data1.index[index_outliers])
```

```python
ax2 = sns.boxplot(y="sepal_width", data=data1_outliers, orient="h", color=sns.color_palette("Set2")[1], linewidth=2.5)
plt.show()
```

# <font color='red'> ** 3) Numeric data discretization **  </font>

```python
data3=df.copy()
```

```python
#X3 = data3.iloc[:, :-1].values
#X3 = data3[['sepal_length','sepal_width', 'petal_length', 'petal_width']].values
X3 = data3[['sepal_length','sepal_width', 'petal_length', 'petal_width']]
```

## Unsupervised: Equal-Width

```python
# transform the dataset with KBinsDiscretizer
from sklearn.preprocessing import KBinsDiscretizer

enc1 = KBinsDiscretizer(n_bins=[3, 4, 2, 2], encode='ordinal', strategy='uniform')
enc1.fit(X3)

X_binned_EW = enc1.transform(X3)
```

```python
X_binned_EW
```

## Unsupervised: Equal-Frequency

```python
# transform the dataset with KBinsDiscretizer
from sklearn.preprocessing import KBinsDiscretizer

enc2 = KBinsDiscretizer(n_bins=[3, 4, 2, 2], encode='ordinal', strategy = 'quantile')
enc2.fit(X3)

X_binned_EF = enc2.transform(X3)
```

```python
X_binned_EF
```

```python
fig =plt.figure(figsize=(20, 20))

plt.subplot(421)
plt.hist(X_binned_EW[:,0])

#plt.xlabel('xlabel',fontsize=22)
#plt.ylabel('ylabel',fontsize=22)
plt.title('Equal-Width -> sepal_length',fontsize=20)

plt.subplot(422)
plt.hist(X_binned_EF[:,0])
plt.title('Equal-Frequency -> sepal_length',fontsize=20)

plt.subplot(423)
plt.hist(X_binned_EW[:,1])
plt.title('Equal-Width -> sepal_width',fontsize=20)

plt.subplot(424)
plt.hist(X_binned_EF[:,1])
plt.title('Equal-Frequency -> sepal_width',fontsize=20)

plt.subplot(425)
plt.hist(X_binned_EW[:,2])
plt.title('Equal-Width -> petal_length',fontsize=20)

plt.subplot(426)
plt.hist(X_binned_EF[:,2])
plt.title('Equal-Frequency -> petal_length',fontsize=20)

plt.subplot(427)
plt.hist(X_binned_EW[:,3])
plt.title('Equal-Width -> petal_width',fontsize=20)

plt.subplot(428)
plt.hist(X_binned_EF[:,3])
plt.title('Equal-Frequency -> petal_width',fontsize=20)

plt.show()
```

# <font color='red'> ** 6) Handling unbalanced data **  </font>

```python
data6=df.copy()
```

```python
X6 = data6[['sepal_length','sepal_width', 'petal_length', 'petal_width']]
y6 = data6['species']
```

```python
y6.head(5)
```

```python
y6.value_counts()
```

```python
data6['species']=y2_encoding
```

```python
data6.head(5)
```

```python
data6=data6.sort_values(by='species', ascending=True)
```

```python
data6.head(5)
```

```python
data6=data6.reset_index(drop=True)
```

```python
data6.head(5)
```

## We should remember that the encoding was: 

### Virginica -> 2
### Versicolor -> 1
### Setosa -> 0

### Lets get an unbalanced data

```python
data6_0=data6[0:50]
data6_1=data6[50:55]
data6_2=data6[100:103]
```

```python
data6_2
```

```python
data6_unbalanced_aux=pd.concat([data6_0,data6_1])
```

```python
data6_unbalanced=pd.concat([data6_unbalanced_aux,data6_2])
```

```python
data6_unbalanced
```

```python
data6_unbalanced['species'].value_counts()
```
